# Machine Learning
**Machine Learning**: subset of artificial intelligence that enables computers to learn from data and improve at tasks, without explicit instructions for every case.

## Supervised Learning
**Supervised Learning**: technique that trains an algorithm to predict outcomes or categorize data using labeled input datasets. *Regression* models predict continuous numerical values and *classification* models predict a discrete category.

There are also *parametric models* with a fixed number of parameters, and *non-parametric models* that have model complexity growing with the data.

**Linear Models**: models that assume the target is a linear combination of the input features. Even if features are transformed (e.g., polynomial features), the model remains linear in the parameters. Examples include:
- `LinearRegression`
- `Ridge`
- `Lasso`
- `ElasticNet`
- `LogisticRegression`
- `SGDRegressor`
- `SGDClassifier`

Use cases include:
- Relationships are approximately linear
- High-dimensional data like text
- Large datasets (especially with SGD)

**Tree-Based Models**: models that recursively split the feature space into regions using decision rules. These are non-linear, capture interactions automatically, and require little preprocessing. Examples include:
- `DecisionTreeClassifier`
- `DecisionTreeRegressor`
- `RandomForestClassifier`
- `RandomForestRegressor`
- `GradientBoostingClassifier`
- `GradientBoostingRegressor`
- `HistGradientBoostingClassifier`
- `HistGradientBoostingRegressor`

Use cases include:
- Mixed numerical and categorical data
- Capturing feature interactions or complex nonlinear patterns
- Tabular structured data

**Boosted models tend to be dominant in machine learning due to their flexibility and strength.**

**Kernel Methods (Support Vector Machines)**: models that find a maximum-margin separating hyperplane. Using the *kernel trick* they can model nonlinear relationships without explicitly transforming features. Examples include:
- `SVC`
- `SVR`
- `LinearSVC`
- `LinearSVR`

Use cases include:
- Medium-sized datasets
- High-dimensional data
- Complex nonlinear boundaries

**Instance-Based Models**: models that store training data and make predictions based on similarity measures. This is not actually "learning", thus this is often called lazy learning. Examples include:
- `KNeighborsClassifer`
- `KNeighborsRegressor`

Use cases include:
- Small datasets
- Simple baseline
- Nonlinear problems

**Probabilistic Models**: models based on probability distributions and Bayes' theorem. Examples include:
- `GaussianNB`
- `MultinomialNB`
- `BernoulliNB`
- `ComplementNB`

Use cases include:
- Text classification
- Spam detection
- Small datasets
- Very high-dimensional sparse data

**Neural Networks (Multi-Layer Perceptron)**: feedforward neural networks trained through backpropagation. Examples include:
- `MLPClassifier`
- `MLPRegressor`

Use cases include:
- Nonlinear problems
- Medium-sized datasets
- When feature engineering is limited

**Ensemble Methods**: some families, particularly tree-based models, include ensembles. These are models that combine multiple base learners to improve performance. The two main types are *bagging* and *boosting*. Examples include:
- `BaggingClassifier`
- `BaggingRegressor`
- `RandomForestClassifier`
- `AdaBoostClassifier`
- `GradientBoostingClassifier`
- `HistGradientBoostingClassifier`

## Unsupervised Learning
**Unsupervised Learning**: technique that analyzes unlabeled datasets to discover hidden patterns, trends, and structures.

**Clustering**: grouping similar samples together. Examples include:
- `KMeans`
- `AgglomerativeClustering`
- `DBSCAN`

Use cases include:
- Customer segmentation
- Image compression

**Dimensionality Reduction**: reducing the number of features while preserving structure. Examples include:
- `PCA`
- `TSNE`

Use cases include:
- Preprocessing
- Visualization
- Noise reduction

**Density Estimation**: estimating the underlying probability distribution that generated the observed data. Examples include:
- `GaussianMixture`

Use cases include:
- Data simulation
- Generative modeling
- Scientific modeling

**Anomaly Detection**: identifying observations that significantly deviate from normal data. Examples include:
- `IsolationForest`

Use cases include:
- Fraud detection
- Network intrusion detection
- Medical abnormalities
- Manufacturing defects
- Outlier removal before training other models
