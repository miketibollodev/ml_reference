## Feature Engineering

We engineer the `HouseAge`, `RemodelAge`, and `GarageAge` for the sake of linear models. A linear model would interpret years incorrectly - for example, a 5-year old house built in 2005 would be seen as much larger than a 35-year old house built in 1975.

```python
target_train = data_train[target_name]
```

```python
data_train["HouseAge"] = data_train["YrSold"] - data_train["YearBuilt"]
data_train["RemodelAge"] = data_train["YrSold"] - data_train["YearRemodAdd"]
data_train["GarageAge"] = data_train["YrSold"] - data_train["GarageYrBlt"]

data_test["HouseAge"] = data_test["YrSold"] - data_test["YearBuilt"]
data_test["RemodelAge"] = data_test["YrSold"] - data_test["YearRemodAdd"]
data_test["GarageAge"] = data_test["YrSold"] - data_test["GarageYrBlt"]
```

```python
data_train["MSSubClass"] = data_train["MSSubClass"].astype("category")
data_train["MoSold"] = data_train["MoSold"].astype("category")
data_train["YrSold"] = data_train["YrSold"].astype("category")

data_test["MSSubClass"] = data_test["MSSubClass"].astype("category")
data_test["MoSold"] = data_test["MoSold"].astype("category")
data_test["YrSold"] = data_test["YrSold"].astype("category")
```

```python
numerical_features = [
    "LotFrontage",
    "LotArea",
    "MasVnrArea",
    "BsmtFinSF1",
    "BsmtFinSF2",
    "BsmtUnfSF",
    "TotalBsmtSF",
    "1stFlrSF",
    "2ndFlrSF",
    "LowQualFinSF",
    "GrLivArea",
    "BsmtFullBath",
    "BsmtHalfBath",
    "FullBath",
    "HalfBath",
    "BedroomAbvGr",
    "KitchenAbvGr",
    "TotRmsAbvGrd",
    "Fireplaces",
    "GarageCars",
    "GarageArea",
    "WoodDeckSF",
    "OpenPorchSF",
    "EnclosedPorch",
    "3SsnPorch",
    "ScreenPorch",
    "PoolArea",
    "MiscVal",
    "HouseAge",
    "RemodelAge",
    "GarageAge"
]

ordinal_features = [
    "LotShape",        # Reg > IR1 > IR2 > IR3
    "LandSlope",       # Gtl > Mod > Sev
    "ExterQual",       # Ex > Gd > TA > Fa > Po
    "ExterCond",
    "BsmtQual",
    "BsmtCond",
    "BsmtExposure",    # Gd > Av > Mn > No
    "BsmtFinType1",    # GLQ > ALQ > BLQ > Rec > LwQ > Unf
    "BsmtFinType2",
    "HeatingQC",
    "KitchenQual",
    "Functional",      # Typ > Min1 > Min2 > Mod > Maj1 > Maj2 > Sev > Sal
    "FireplaceQu",
    "GarageFinish",    # Fin > RFn > Unf
    "GarageQual",
    "GarageCond",
    "PavedDrive",      # Y > P > N
    "PoolQC",
    "Fence"            # GdPrv > MnPrv > GdWo > MnWw
]

nominal_features = [
    "MSSubClass",
    "MSZoning",
    "Street",
    "Alley",
    "LandContour",
    "Utilities",
    "LotConfig",
    "Neighborhood",
    "Condition1",
    "Condition2",
    "BldgType",
    "HouseStyle",
    "RoofStyle",
    "RoofMatl",
    "Exterior1st",
    "Exterior2nd",
    "MasVnrType",
    "Foundation",
    "Heating",
    "CentralAir",
    "Electrical",
    "GarageType",
    "MiscFeature",
    "SaleType",
    "SaleCondition"
]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer

numerical_pipeline = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)

nominal_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)

ordinal_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
)

preprocessor = make_column_transformer(
    (numerical_pipeline, numerical_features),
    (nominal_pipeline, nominal_features),
    (ordinal_pipeline, ordinal_features),  
    remainder="drop"
)
```

## Linear Model

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import RandomizedSearchCV, KFold

cv = KFold(n_splits=5, shuffle=True, random_state=0)

model = ElasticNetCV(
    l1_ratio=[0.1, 0.3, 0.5, 0.7, 0.9, 1.0],
    alphas=None,
    cv=5,
    max_iter=10_000
)

pipeline = make_pipeline(
    preprocessor,
    model
)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error

rmse_scorer = make_scorer(
    lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)),
    greater_is_better=False
)

score = cross_val_score(
    pipeline, 
    data_train, 
    target_train, 
    cv=cv, 
    scoring=rmse_scorer
)

print(f"Mean RMSE: {-score.mean():.2f} +/- {score.std():.2f}")
```

## Fit Model

```python
pipeline.fit(data_train, target_train)
```

```python
predictions = pipeline.predict(data_test)

output = pd.DataFrame({
    identifier_name: data_test[identifier_name],
    target_name: predictions
})

output.to_csv("submission.csv", index=False)
```
